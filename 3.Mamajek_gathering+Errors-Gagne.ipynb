```python
#!/usr/bin/env python
import numpy as np
import pylab as plt
import scipy
import scipy.interpolate
from scipy.interpolate import UnivariateSpline
import os
import my_io
import matplotlib.patches as patches


#Lectura de las tablas desde archivos de texto.
#filename1 es la tabla a partir de la cual hacer la interpolacion, de Mamajek
#filename2 es la tabla con los objetos miembros a los YNMG                      
filename1='referencia_mamajek.txt_number_code'
filename2='YNMG_8402_Ext.csv'
#filename2='grupo_control.2Av'
tabla=scipy.genfromtxt(filename1,skip_header=1,dtype=float)
catalogo=scipy.genfromtxt(filename2,skip_header=1,delimiter=",",dtype='float',filling_values='-0')

#Lectura de las variables de la tabla de Mamajek
SpT=tabla[:,0].astype(float) #Tipo Espectral
Teff=tabla[:,2].astype(float) #Temperatura efectiva
BCv=tabla[:,3].astype(float) #Correccion bolometrica
J_H=tabla[:,13].astype(float) #color J-H
H_K=tabla[:,14].astype(float) #color H-Ks
J_K=J_H+H_K #Construccion del color J-Ks 

#Lectura de variables del catalogo que contiene informacion para los miembros de las YNMG
Jmag=catalogo[:,9].astype(float)  #Magnitudes 2MASS
Hmag=catalogo[:,10].astype(float) 
Kmag=catalogo[:,11].astype(float)

e_Jmag=catalogo[:,12].astype(float)  #Errores en magnitudes 2MASS
e_Hmag=catalogo[:,13].astype(float) 
e_Kmag=catalogo[:,14].astype(float)

SpT_code=catalogo[:,30].astype(float)
SpT_Gagne=catalogo[:,36].astype(float)
EBV_Sh=catalogo[:,37].astype(float) #Exceso E(B-V) Schlegel
EBV_Gr=catalogo[:,38].astype(float) #Exceso E(B-V) Green

#Construccion de colores observados, a partir de magnitudes 2MASS
jmag_hmag=Jmag-Hmag
jmag_kmag=Jmag-Kmag
hmag_kmag=Hmag-Kmag
 
Rv=3.1 # Ley de extincion asumida Rieke & Lebofsky 1985
e_Rv=0.03 # error Ley de extincion
# relaciones entre absorciones en distintas bandas (lambda y V)
c_Jv=0.282 # A_J/A_V
c_Hv=0.175 # A_H/A_V
c_Kv=0.112 # A_K/A_V
# constantes de color
c_JH=0.107 # E(J-H)=A_J-A+H => Av (0.282-0.175) => Av(0.107) JH
c_JK=0.17  # JK
c_HK=0.063 # HK

Av_Gr=EBV_Gr*Rv #Extincion Green
Av_Sch=EBV_Sh*Rv #Extincion Schlegel

print min(SpT_code), max(SpT_code)
```

```python
#Funcion de seleccion para crear columna en la que se asume el tipo espectral proveeido por Gagne
#si disponible o el de SIMBAD, o 0. 

spt_code=np.array([])
n = len(SpT_code)

for i in range(n):
    if SpT_Gagne[i] != -0:
        sPT=SpT_Gagne[i]
        spt_code=np.append(spt_code,sPT)
    elif SpT_code[i] != -0:
        sPT=SpT_code[i]
        spt_code=np.append(spt_code,sPT)
    else:
        sPT=0
        spt_code=np.append(spt_code,sPT)

#print spt_code[614], SpT_Gagne[614], SpT_code[614]
#print spt_code[783], SpT_Gagne[783], SpT_code[783]
#print spt_code[1044], SpT_Gagne[1044], SpT_code[1044]
##print spt_code[1513], SpT_Gagne[1513], SpT_code[1513]
#print spt_code[5614], SpT_Gagne[5614], SpT_code[5614]
#print len(spt_code)
```

```python
#Para la interpolacion de color intrinseco a partir del tipo espectral, ordeno y aplico el orden
#del tipo espectral a las otras variables
SPT_sorted=SpT.argsort() 
SPT=SpT[SPT_sorted]

teff_spt=Teff[SPT_sorted]
BCv_spt=BCv[SPT_sorted]
```

```python
#Restriccion de la tabla de Mamajek para el color J-H
j_h_mask=(J_H!=-99)&(Teff>1700)    #Remueve valores ficticios y enanas tipo T (azulamiento)
#Aplicacion de la mascara
JHr=J_H[j_h_mask]
teff_jhr=Teff[j_h_mask]
BCv_jhr=BCv[j_h_mask]
SpT_jhr=SpT[j_h_mask]
#Para interpolar es necesario ordenar la variable dependiente respecto al orden de la variable
#independiente, en este caso del color J-H. Esto NO soluciona el problema de las MULTIEVALUACIONES.
#Si no se ordena, no se puede interpolar. 
j_h_sorted=JHr.argsort()
j_h_referencia=JHr[j_h_sorted]
teff_jh=teff_jhr[j_h_sorted]
BCv_jh=BCv_jhr[j_h_sorted]
SpT_jh=SpT_jhr[j_h_sorted]

#Para la interpolacion en la tabla de Mamajek del color J-H intrinseco a partir del tipo espectral aplico el orden del SpT
jh_spt=J_H[SPT_sorted]
```

```python
#Restriccion de la tabla de Mamajek para el color J-K
j_k_mask=(J_K!=-99)&(Teff>1590)   #Remueve valores ficticios y enanas tipo T (azulamiento)
#Aplicacion de la mascara
JKr=J_K[j_k_mask]
teff_jkr=Teff[j_k_mask]
BCv_jkr=BCv[j_k_mask]
SpT_jkr=SpT[j_k_mask]
#Para interpolar es necesario ordenar la variable dependiente respecto al orden de la variable
#independiente, orden del color J-K. Esto no soluciona el problema de las multievaluaciones.
#Si no se ordena, no se puede interpolar.
j_k_sorted=JKr.argsort()
j_k_referencia=JKr[j_k_sorted]
teff_jk=teff_jkr[j_k_sorted]
BCv_jk=BCv_jkr[j_k_sorted]
SpT_jk=SpT_jkr[j_k_sorted]

#Para la interpolacion en la tabla de Mamajek del color J-K intrinseco a partir del tipo espectral aplico el orden del SpT
jk_spt=J_K[SPT_sorted]
```

```python
#Restriccion de la tabla de Mamajek para el color H-K
h_k_mask=(H_K!=-99)&(Teff>1590)#Remueve valores ficticios y enanas tipo T (azulamiento)
#Aplicacion de la mascara
HKr=H_K[h_k_mask]
teff_hkr=Teff[h_k_mask]
BCv_hkr=BCv[h_k_mask]
SpT_hkr=SpT[h_k_mask]
#Para interpolar es necesario ordenar la variable dependiente respecto al orden de la variable
#independiente, orden del color H-K. Esto no soluciona el problema de las multievaluaciones.
#Si no se ordena, no se puede interpolar.
h_k_sorted=HKr.argsort()
h_k_referencia=HKr[h_k_sorted]
teff_hk=teff_hkr[h_k_sorted]
BCv_hk=BCv_hkr[h_k_sorted]
SpT_hk=SpT_hkr[h_k_sorted]

#Para la interpolacion del color H-K intrinseco a partir del tipo espectral aplico el orden del SpT
hk_spt=H_K[SPT_sorted]
```

```python
#Interpolaciones

#Para SIMBAD, obtengo colores intrinsecos JH, JK y HK; Teff y BCv a partir del tipo espectral *Mamajek

#Interpolacion lineal del color intrinseco J-H a partir del tipo espectral de SIMBAD
fJ_H=scipy.interpolate.interp1d(SPT,jh_spt,kind='linear',bounds_error=False,fill_value='NaN')
#Interpolacion lineal del color J-K a partir del tipo espectral de SIMBAD
fJ_K=scipy.interpolate.interp1d(SPT,jk_spt,kind='linear',bounds_error=False,fill_value='NaN')
#Interpolacion lineal del color H-K a partir del tipo espectral de SIMBAD
fH_K=scipy.interpolate.interp1d(SPT,hk_spt,kind='linear',bounds_error=False,fill_value='NaN')

#Interpolacion lineal de la temperatura effectiva a partir del tipo espectral de SIMBAD
f_teff_spt=S=scipy.interpolate.interp1d(SPT,teff_spt,kind='linear',bounds_error=False,fill_value='NaN')
#Interpolacion lineal de la correccion bolometrica a partir del tipo espectral de SIMBAD
f_BCv_spt=S=scipy.interpolate.interp1d(SPT,BCv_spt,kind='linear',bounds_error=False,fill_value='NaN')
```

```python
#Interpolaciones, a partir de los colores obtengo SpT, Teff y BCv *Mamajek

#Interpolaciones en la tabla de Mamajek, a partir de J-H intrinseco se obtiene SpT, Teff, BCv
f_jh_teff=scipy.interpolate.interp1d(j_h_referencia,teff_jh,kind='linear',bounds_error=False,fill_value='NaN')
f_jh_bcv=scipy.interpolate.interp1d(j_h_referencia,BCv_jh,kind='linear',bounds_error=False,fill_value='NaN')
f_jh_spt=scipy.interpolate.interp1d(j_h_referencia,SpT_jh,kind='linear',bounds_error=False,fill_value='NaN')

#Interpolaciones en la tabla de Mamajek, a partir de J-K intrinseco se obtiene SpT, Teff, BCv
f_jk_teff=scipy.interpolate.interp1d(j_k_referencia,teff_jk,kind='linear',bounds_error=False,fill_value='NaN')
f_jk_bcv=scipy.interpolate.interp1d(j_k_referencia,BCv_jk,kind='linear',bounds_error=False,fill_value='NaN')
f_jk_spt=scipy.interpolate.interp1d(j_k_referencia,SpT_jk,kind='linear',bounds_error=False,fill_value='NaN')

#Interpolaciones en la tabla de Mamajek, a partir de H-K intrinseco se obtiene SpT, Teff, BCv
f_hk_teff=scipy.interpolate.interp1d(h_k_referencia,teff_hk,kind='linear',bounds_error=False,fill_value='NaN')
f_hk_bcv=scipy.interpolate.interp1d(h_k_referencia,BCv_hk,kind='linear',bounds_error=False,fill_value='NaN')
f_hk_spt=scipy.interpolate.interp1d(h_k_referencia,SpT_hk,kind='linear',bounds_error=False,fill_value='NaN')
```
